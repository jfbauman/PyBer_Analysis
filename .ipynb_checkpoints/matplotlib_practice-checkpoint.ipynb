```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() "Does the same as above code"
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() "Does the same as above code"
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Bubble Chart using MATLAB Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis, y_axis, c='blue', s=y_axis)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Bubble Chart using MATLAB Method
#list comprehension= easier than for loop

# Import dependencies.
import matplotlib.pyplot as plt
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis, y_axis, c='blue', s=y_axis)
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Bubble Chart using Object-Oriented Interface
# Import dependencies.
import matplotlib.pyplot as plt
# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, c='skyblue', marker='o', edgecolors="black", alpha=0.25, s=100, label='Boston')

plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
ax.legend()
```

```python
# Create a Pie Chart Using the MATLAB Method
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# exploding the 7th label 'July'
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#increase the size of the pie chart 
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors, 
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
#Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# exploding the 7th label 'July'
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,labels=x_axis)
#increase the size of the pie chart 
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors, 
        labels=x_axis, 
        autopct='%.1f%%', shadow=True, startangle= 90, counterclock=False)

plt.show()
```
